```python
print(text_rank_list[4])
```

```python
dictionary = corpora.Dictionary(text_rank_list) 
#dictionary.filter_extremes(no_below=3)
corpus = [dictionary.doc2bow(sentence) for sentence in text_rank_list]

lda = models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=15)

for topic in lda.print_topics(num_topics=15, num_words=10):
            #print(uid)
            print(topic[0])
            print(topic[1])
            #lda_1 = []
            #lda_1.append(uid)
            #lda_1.append(topic[0])
            #lda_1.append(topic[1])

            #csv_write.writerow(lda_1)
```

```python
vis = pyLDAvis.gensim.prepare(topic_model=lda, corpus=corpus, dictionary=dictionary,sort_topics=False)
pyLDAvis.enable_notebook()
pyLDAvis.display(vis)
```

```python
lda.save('weibo_lda.model')
```

```python
pyLDAvis.save_html(vis, 'pyLDAvis.html')
```

```python
for topic in lda.print_topics(num_topics=15, num_words=20):
            #print(uid)
            print(topic[0])
            print(topic[1])
```

```python
topic_index = []
for i in range(0,len(corpus)) :
    #print(corpus[i])
    
    max_topic = sorted(lda[corpus[i]], reverse=True, key = lambda x: x[1])[0]
    if max_topic[1] < 0.1 :
        topic_index.append(-1)
    else :
        topic_index.append(max_topic[0])
    
print(topic_index)    
```

```python
df1['topic_num'] = topic_index
print(df1.iloc[18])
```
