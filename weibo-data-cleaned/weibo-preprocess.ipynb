```python
def remove_du (df0) :
    df0.drop(df0.index[df0['tool'] == '微公益'], inplace=True)
    #df1.drop(df1.index[df1['tool'] == '微博视频号'], inplace=True)
    df0['text'] = df0['text'].map(lambda x: re.sub('L.+的微博视频', '',x))

    df0.drop_duplicates(subset='text', keep="last",inplace = True)
    return df0

    
df2 = pd.read_csv('./result1/%23山西暴雨%23/%23山西暴雨%23.csv')
df3 = pd.read_csv('./result1/山西 暴雨/山西 暴雨.csv')
df4 = pd.read_csv('./result1/山西 伤亡/山西 伤亡.csv')
df5 = pd.read_csv('./result1/山西 坍塌/山西 坍塌.csv')
df6 = pd.read_csv('./result1/山西 救援/山西 救援.csv')
df7 = pd.read_csv('./result1/山西 洪水/山西 洪水.csv')
df8 = pd.read_csv('./结果文件/山西 暴雨/山西 暴雨.csv')

print(len(remove_du (df2).index))
print(len(remove_du (df3).index))
print(len(remove_du (df4).index))
print(len(remove_du (df5).index))
print(len(remove_du (df6).index))
print(len(remove_du (df7).index))
print(len(remove_du (df8).index))

df1 = pd.concat([remove_du (df2), remove_du(df3),remove_du(df4),remove_du(df5),remove_du(df6),remove_du(df7),remove_du(df8)],ignore_index=True)

print(len(df1.index))
df1.drop_duplicates(subset='text',keep = 'last',inplace =True)
print(len(df1.index))

```

```python
def remove_topic(text) :
    index = [i for i,val in enumerate(list(text)) if val=='#']
    
    yy = text
    ii = 0
    for i in range(0, len(index)) :
        if i % 2 == 0 :
            
            ii = i
        else :
            
            yy = re.sub(text[index[ii]:index[i]+1],'',yy)
            
    return yy

def remove_at (text, at_user) :
   # print(at_user)
    if not pd.isnull(at_user) :
        #print(False)
        users = at_user.split(',')
        at_users = list(map(lambda x: '@'+x , users))
        at_users_regex = re.compile('|'.join(map(re.escape, at_users)))
        return re.sub(at_users_regex, '', text)
    else :
        return text

def pre_p(text) :
    pre_text = preprocess.preprocess(text, keep_stop_word=False)
    pre_text_sub = re.sub(r'[^\u4e00-\u9fff\s]+','',pre_text)
    return pre_text_sub.split()


    
df1['text'] = df1['text'].map(lambda x: remove_topic(x))

df1['text_pre'] = df1.apply(lambda row: remove_at(row.text,row.at_user), axis=1)
df1.drop(df1.index[df1['text_pre'] == ''], inplace=True)
df1['text_pre_list'] = df1['text_pre'].map(lambda x: pre_p(x))
df1['date'] = pd.to_datetime(df1['date'])

print(len(df1.index))
```
